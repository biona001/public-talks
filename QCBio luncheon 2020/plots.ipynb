# Illustration of p values

```julia
using Plots
Plots.scalefontsizes(2)
```

```julia
function simulate_and_plot(n)
    x = rand(0:2, n)
#     y = 100 .+ 10 .* randn(n)  #no effect
    y = 100 .+ 10x .+ 5randn(n)
    return scatter(x, y, xlabel="Number of A2 alleles at a cite", ylabel = "Average blood pressure", label="blood pressure")
end
```

```julia
n = 100
p = simulate_and_plot(n)
p
```

```julia
savefig(p, "plot2.pdf")
```

# Easy to be mislead when one doesnot control for confounding

```julia
male_age = collect(30:5:60)
male_blood_pressure = collect(80:10:140)

female_age = collect(10:5:40)
female_blood_pressure = collect(80:10:140)

age = [male_age; female_age]
blood_pressure = [male_blood_pressure; female_blood_pressure]
```

```julia
scatter(age, blood_pressure, xlabel="age", 
        ylabel = "Average blood pressure", 
        label="blood pressure", 
        legend=:topleft)
```
